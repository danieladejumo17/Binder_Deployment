```python
# from fastbook import *
from fastai.vision.all import *
from fastai.vision.widgets import *
```

```python
!cp "/content/drive/MyDrive/Python/Machine Learning/FastAI Course 22/Pass 2/bird_or_forest.pkl" .
```

```python
# load learner
model_path = "bird_or_forest.pkl"
learner = load_learner(model_path)
```

```python
# Upload Button
upload_btn = widgets.FileUpload()
```

```python
# Output widget used by Viola
img_output = widgets.Output()
```

```python
# Label Output
pred_label = widgets.Label()
```

```python
# Button to trigger classification and callback function

def classify_image(chng):
  img = PILImage.create(upload_btn.data[-1])
  img_output.clear_output()
  with img_output: display(img.to_thumb(192, 192))
  pred, idx, probs = learner.predict(img)
  pred_label.value = f"Prediction: {pred}; Probability: {(probs[idx]*100):.1f}%"

classify_btn = widgets.Button(description="Classify")
classify_btn.on_click(classify_image)
```

```python
# Organize the widgets in a layout

VBox([widgets.Label("Select your image"),
     upload_btn,
     classify_btn,
     img_output,
     pred_label])
```
